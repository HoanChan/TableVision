```python
import os
import xml.etree.ElementTree as ET
class BoundingBox:
    def __init__(self, xmin, ymin, xmax, ymax):
        self.xmin = xmin
        self.ymin = ymin
        self.xmax = xmax
        self.ymax = ymax

class Size:
    def __init__(self, width, height, depth):
        self.width = width
        self.height = height
        self.depth = depth

class PVOC:
    def __init__(self, path, name, pose, truncated, difficult, occluded, xmin, ymin, xmax, ymax, width, height, depth, segmented):
        self.path = path
        self.name = name
        self.pose = pose
        self.truncated = truncated
        self.difficult = difficult
        self.occluded = occluded
        self.bndbox = BoundingBox( xmin = xmin, ymin = ymin, xmax = xmax, ymax = ymax )
        self.size = Size( width = width, height = height, depth = depth)
        self.segmented = segmented

def read_xml_files(folder_path):
    # Đường dẫn đến thư mục chứa các file XML
    path = folder_path

    # Kiểm tra xem đường dẫn thư mục tồn tại hay không
    if not os.path.exists(path):
        print(f"Thư mục '{path}' không tồn tại.")
        return []

    # Lấy danh sách các file XML trong thư mục
    xml_files = [f for f in os.listdir(path) if f.endswith('.xml')]

    objects = []  # Danh sách các đối tượng từ các file XML

    # Đọc nội dung của từng file XML và chuyển đổi thành đối tượng
    for xml_file in xml_files:
        # Đường dẫn đến file XML
        xml_path = os.path.join(path, xml_file)

        # Đọc nội dung file XML và tạo cây phân tích cú pháp
        tree = ET.parse(xml_path)
        root = tree.getroot()

        # Tạo đối tượng PVOC từ thông tin thu thập được
        obj = PVOC(
            path = "/content/pubtables-1m/PubTables-1M-Detection/images/" + root.find('filename').text,
            name = root.find('object/name').text,
            pose = root.find('object/pose').text,
            truncated = root.find('object/truncated').text,
            difficult = root.find('object/difficult').text,
            occluded = root.find('object/occluded').text,
            xmin = root.find('object/bndbox/xmin').text,
            ymin = root.find('object/bndbox/ymin').text,
            xmax = root.find('object/bndbox/xmax').text,
            ymax = root.find('object/bndbox/ymax').text,
            width = root.find('size/width').text,
            height = root.find('size/height').text,
            depth = root.find('size/depth').text,
            segmented = root.find('segmented').text
        )

        objects.append(obj)

    return objects


# Thay đổi folder_path thành đường dẫn thư mục chứa các file XML
folder_path = "/content/pubtables-1m/PubTables-1M-Detection/train"
objects = read_xml_files(folder_path)

# In thông tin các đối tượng
for obj in objects:
    print("Path:", obj.path)
    print("Name:", obj.name)
    print("Pose:", obj.pose)
    print("Truncated:", obj.truncated)
    print("Difficult:", obj.difficult)
    print("Occluded:", obj.occluded)
    print("Bounding Box:")
    print("  xmin:", obj.bndbox.xmin)
    print("  ymin:", obj.bndbox.ymin)
    print("  xmax:", obj.bndbox.xmax)
    print("  ymax:", obj.bndbox.ymax)
    print("Size:")
    print("  Width:", obj.size.width)
    print("  Height:", obj.size.height)
    print("  Depth:", obj.size.depth)
    print("Segmented:", obj.segmented)
    print("-----------------------------")
    break;
```

```python
import cv2
import matplotlib.pyplot as plt

def draw_bounding_boxes(images, indices):
    num_images = len(indices)
    fig, axes = plt.subplots(1, num_images, figsize=(15, 5))

    for i, index in enumerate(indices):
        image_info = images[index - 1]
        image_path = image_info.path
        image = cv2.imread(image_path)

        # Lấy thông tin bounding box
        xmin = float(image_info.bndbox.xmin)
        ymin = float(image_info.bndbox.ymin)
        xmax = float(image_info.bndbox.xmax)
        ymax = float(image_info.bndbox.ymax)

        # Vẽ bounding box lên ảnh
        cv2.rectangle(image, (int(xmin), int(ymin)), (int(xmax), int(ymax)), (0, 255, 0), 2)

        # Hiển thị ảnh với bounding box trên subplot tương ứng
        axes[i].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        axes[i].axis('off')

    plt.show()
```

```python
import random
indices_to_draw = random.choices(range(len(objects)), k= 5)
print(indices_to_draw)
draw_bounding_boxes(objects, indices_to_draw)
```

```python
import os
import xml.etree.ElementTree as ET

class BoundingBox:
    def __init__(self, name, pose, truncated, difficult, occluded, xmin, ymin, xmax, ymax):
        self.name = name
        self.pose = pose
        self.truncated = truncated
        self.difficult = difficult
        self.occluded = occluded
        self.bbox = [xmin, ymin, xmax, ymax]

class Size:
    def __init__(self, width, height, depth):
        self.width = width
        self.height = height
        self.depth = depth

class PVOC:
    def __init__(self, path, size, segmented, bndboxes):
        self.path = path
        self.size = size
        self.segmented = segmented
        self.bndboxes = bndboxes

def readXML(xml_path):
    # Đọc nội dung file XML và tạo cây phân tích cú pháp
    tree = ET.parse(xml_path)
    root = tree.getroot()

    # Lấy thông tin chung
    size = Size(
        width=root.find('size/width').text,
        height=root.find('size/height').text,
        depth=root.find('size/depth').text
    )
    segmented = root.find('segmented').text

    # Tạo danh sách bounding boxes
    bndboxes = []

    for object_ in root.iter('object'):
        name = object_.find('name').text
        pose = object_.find('pose').text
        truncated = object_.find('truncated').text
        difficult = object_.find('difficult').text
        occluded = object_.find('occluded').text
        for box in object_.findall("bndbox"):
            xmin = float(box.find('xmin').text)
            ymin = float(box.find('ymin').text)
            xmax = float(box.find('xmax').text)
            ymax = float(box.find('ymax').text)
        bndboxes.append(BoundingBox(name, pose, truncated, difficult, occluded, xmin, ymin, xmax, ymax))

    # Tạo đối tượng PVOC
    obj = PVOC(
        path=root.find('filename').text,
        bndboxes=bndboxes,
        size=size,
        segmented=segmented
    )
    return obj

obj = readXML(xml_path)

print("Path:", obj.path) 
print("Size:")
print("  Width:", obj.size.width)
print("  Height:", obj.size.height)
print("  Depth:", obj.size.depth)
print("Segmented:", obj.segmented)
print("Bounding Boxes:")
for bndbox, i in zip(obj.bndboxes, range(len(obj.bndboxes))):
    print('  box', i + 1, ":", bndbox.name, bndbox.pose, bndbox.truncated, bndbox.difficult, bndbox.occluded, bndbox.bbox)
```
