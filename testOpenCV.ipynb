```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Tạo ảnh màu đen
img = np.zeros((500, 500, 3), dtype=np.uint8)

# tạo point1 và point2 ngẫu nhiên
point1 = np.random.randint(0, 500, 2)
point2 = np.random.randint(0, 500, 2)

# Tạo một mảng chứa tọa độ các điểm trên đường thẳng
line_pts = np.linspace(point1, point2, 100).astype(int)

# Vẽ đường thẳng

for pt in line_pts[::2]:
    cv2.circle(img, tuple(pt), 1, (255, 255, 255), 1)

from utils.point import is_have_line

# Thiết lập màu của đường thẳng thành trắng
print(is_have_line((point1[0], point1[1]), (point2[0], point2[1]), img, 0.1))

plt.imshow(img)
```
